```python
#Order Trends Analysis
#Total Monthly Orders: Pre-Crisis vs Crisis vs Recovery
import matplotlib.pyplot as plt
import seaborn as sns
# Example dataframe: monthly_orders
import pandas as pd
monthly_orders = pd.DataFrame({
    'phase': ['pre-crisis', 'crisis', 'recovery'],
    'total_orders': [113111, 35085, 970],
    'month': ['Jan–May', 'Jun–Sep', 'Oct'] 
})
# Markdown description
from IPython.display import display, Markdown
display(Markdown("### Total Monthly Orders"))
display(Markdown("Total Orders dropped by **68.98%** during the crisis period."))
# Line plot
plt.figure(figsize=(10,5))
for phase in monthly_orders['phase'].unique():
    subset = monthly_orders[monthly_orders['phase']==phase]
    plt.plot(subset['month'].astype(str), subset['total_orders'], marker='o', label=phase)
plt.xticks(rotation=45)
plt.ylabel('Total Orders')
plt.title('Monthly Orders: Pre-Crisis vs Crisis vs Recovery')
plt.legend()
plt.show()
```

### Monthly Orders Trend: Severe Decline During Crisis

QuickBite experienced an **extreme collapse in order volume** during the crisis period.

#### Key Numbers
| Phase       | Total Orders |
|-------------|--------------|
| Pre-Crisis  | **113,111**  |
| Crisis      | **35,085**   |
| Recovery    | **970**      |

#### What This Means
- Orders **dropped by 68.98%** during the crisis compared to pre-crisis levels.
- This is a **catastrophic fall**, signaling:
  - Massive customer drop-off  
  - Loss of trust after food safety incidents  
  - Delivery outages during monsoon season  
  - Competitors capturing QuickBite’s market share  

#### Interpretation
- The crisis caused an **immediate operational and reputational shock**.  
- The recovery phase shows **almost no rebound** (only 970 orders), meaning customer confidence remains extremely low.

#### Business Impact
- Revenue was directly hit due to falling order counts.
- This trend alone justifies a **full-scale recovery and trust-building plan**.

### Delivery SLA Performance: Pre-Crisis vs Crisis vs Recovery

The delivery-time metrics show a clear operational breakdown during the crisis:

| Phase       | Avg Delivery Time | Avg SLA Time | SLA Compliance |
|-------------|-------------------|--------------|----------------|
| Pre-crisis  | **39.52 min**     | 37.50 min    | **43.60%**     |
| Crisis      | **60.13 min**     | 42.52 min    | **12.21%**     |
| Recovery    | **44.52 min**     | 38.95 min    | **35.05%**     |

#### Key Insights
- **Average delivery time increased by +20.6 minutes during the crisis**, confirming major delays.
- **SLA compliance collapsed from 43.6% → 12.2%**, a **72% performance deterioration**.
- SLA improved during recovery (35%), but **is still 20% below pre-crisis levels**.
- SLA delays closely align with customer rating drops and order cancellations observed in other metrics.

#### Business Implications
- The crisis caused widespread operational failures leading to slow deliveries.
- Reduced SLA compliance likely triggered **negative reviews**, **order cancellations**, and **customer churn**.
- Even in the recovery phase, performance is **not fully restored**, suggesting:
  - Incomplete staffing or partner recovery  
  - Traffic/logistical bottlenecks  
  - Restaurant capacity issues  

#### Actionable Recommendations
- Prioritize **high-volume cities** with worst SLA drops.
- Reintroduce/scale **delivery partner incentives** to improve supply levels.
- Implement **dynamic load balancing** to shift orders away from overloaded restaurants.
- Monitor SLA weekly until compliance stabilizes back to **>40%**.
